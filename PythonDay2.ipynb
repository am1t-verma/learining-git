## Arithmetic Operators

```python
num1 = 10
num2 = 3
print("Addition: ",num1+num2)      #To add we use + operator
print("Subtraction: ",num1-num2)     # To subract we use - operator
print("Multiplication: ",num1*num2)  # To multiply we use * operator
print("Division: ",num1/num2)       # To divide to get the quotient part we use / operator
print("Remainder: ", num1%num2)    # To get the remainder we use % operator
print("Floor Division: ",num1//num2)    #To get the floor division(The quotient before decimal) we use // operator
print("Exponential: ", num1**num2)       # ** used to get the exponential
```

# Comparision Operators

```python
a = 50
b = 30
print(a>b)     # Whether a is greater than b   >             TRUE
print(a<b)     # Whether a is smaller than b   <             FALSE
print(a==b)    # To check a is equal to b  ==                FALSE
print(a!=b)    # To check a is not equal to b      !=        TRUE
print(a>=b)    # To check wheter a is greater than or equal to b     >=
print(a<=b)    # To check whether a is less than or equal to b      <=
```
