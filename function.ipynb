```python
def calsi():
    a = int(input())
    b = int(input())
    print("Add  → addition")
    print("Sub  → subtraction")
    print("Div  → division")
    print("Multi → multiplication")
    op = input()
    op = op.upper()
    print("\n")
    if op == "ADD":
        print('Addition of Input NO :',a + b)
    elif op == "SUB":
        print('Subtraction of Input NO :',a - b)
    elif op == "MULTI":
        print('Multiplication of Input NO :',a * b)
    elif op == "DIV":
        print('Division of Input NO :',a / b)
    else:
        print("Invalid operator")

calsi()
```

```python
def details(name, age, roll_no):
    inp  = int(input('enter no 1 for name, 2 for age, 3 for roll-no'))

    if inp == 1:
        return name
    
    elif inp == 2:
        return age
    elif inp == 3:
        return roll_no
    
    else :
        return print('input is incorrect')
    print('this will never print')
```

```python
details('alisha', 21, 23)
```
