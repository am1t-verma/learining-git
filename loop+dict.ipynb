```python
# Print numbers from 1 to N using a loop.
N = int(input('enter a valid no.'))
print(f'Your No. Is :{N}')
a = 1
while N>=a:
    print(a)
    a = a+1
```

```python
# Print all even numbers from 1 to N using a loop.

N = int(input('enter a valid no.'))
a = 2
while N>=a:
    print(a)
    a = a+2
```

```python
# Print the sum of first N natural numbers using a loop.
sum  = 0
N = int(input('enter a valid no.'))
for i in range(1, N+1):
    sum = sum + i



print(sum)
```

```python
# Count how many digits are in a given number using a loop.

Input= int(input('enter a valid no for count the integers'))
if Input == 0:
    print(1)
else:
    N = 0
    while Input!=0:
        Input = Input//10
        N = N + 1
        # print(Input)

    print(N)
```

```python
# Reverse a given number using a loop.
# Output: 54321
Input= 12345
rev = 0
l  = []
while Input!=0:

    last = Input%10
    rev = rev * 10 + last
    l.append(last)
    Input = Input//10

print(l)
print(rev)
```
