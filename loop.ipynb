```python
# Print numbers from 1 to N using a loop.
N = int(input('enter a valid no.'))
print(f'Your No. Is :{N}')
a = 1
while N>=a:
    print(a)
    a = a+1
```

```python
# Print all even numbers from 1 to N using a loop.

N = int(input('enter a valid no.'))
a = 2
while N>=a:
    print(a)
    a = a+2
```

```python
# Print the sum of first N natural numbers using a loop.
sum  = 0
N = int(input('enter a valid no.'))
for i in range(1, N+1):
    sum = sum + i



print(sum)
```
